```python
import time
distance = got.read_distance_data(21)

# --- Step 1: Define the zone rules ---
zones = {
    "GO":   {"min": 80, "color": "green"},
    "SLOW": {"min": 30, "color": "blue"},
    "STOP": {"min": 0,  "color": "red"}
}

# --- Step 2: Main loop ---
while True:
    distance = got.read_distance_data(21)
    print("Distance:", distance)

    # --- Step 3: Decide which zone we are in ---
    if distance > zones["GO"]["min"]:
        zone = "GO"
    elif distance > zones["SLOW"]["min"]:
        zone = "SLOW"
    else:
        zone = "STOP"

    # --- Step 4: Use dictionary values directly to control LED ---
    color = zones[zone]["color"]

    if color == "red":
        got.show_light_rgb_effect(225, 0, 0, 3)
    elif color == "green":
        got.show_light_rgb_effect(0, 225, 0, 3)
    elif color == "blue":
        got.show_light_rgb_effect(0, 0, 225, 3)

    # Optional: stop when too close
    if zone == "STOP":
        break

    time.sleep(0.2)
```

```python
commands = {
    "red": (255, 0, 0),
    "green": (0, 255, 0),
    "blue": (0, 0, 255),
    }

print("Please say a color: red, green, or blue.")

while True:
    asr_result = got.start_audio_asr()
    if not asr_result:
        continue

    got.start_audio_nlp(asr_result, False)
    print(asr_result)

    command = asr_result.lower()
    if "stop" in command:
        got.turn_off_lights()
        print("It's nice talking to you, goodbye!")
        break

    for name,(r,g,b) in commands.items():
        if name in command:
            got.show_light_rgb_effect(r,g,b,3)
            print(f"The light is changed to {name}")
            break
```

1. voice recognition and chassis control

```python
import time
movements = {
    "forward":(0,10,0),
    "backward":(0,-10,0),
    "left":(-10,0,0),
    "right":(10,0,0),
    "rotate":(0,0,10),   
}
print("Please tell me how to move.")

while True:
    asr_result = got.start_audio_asr()
    if not asr_result:
        time.sleep(0.1)
        continue

    command = asr_result.lower().strip()

    for key,(x,y,z) in movements.items():
        if key in command:
            got.mecanum_move_xyz(x,y,z)
            print(f"I am moving {key}")
            break

    if "stop" in command:
        got.mecanum_stop()
        print("Game over!")
        break
```

```python
def AP_recognition():
    # Step 1 AP recognition
    import time
    got.load_models(["apriltag_qrcode"])
    AP_info = got.get_apriltag_total_info()
    
    while True:
        AP_info = got.get_apriltag_total_info()
        if not AP_info:
            got.mecanum_move_xyz(0,0,20)
            time.sleep(0.1)
        else:
            got.mecanum_stop()
            break
            
    print("Apriltag recognized.")  
    
    # Step 2 AP centralization
    while True:
        AP_info = got.get_apriltag_total_info()
        x_coordinate = AP_info[0][1]
        if x_coordinate<320:
            got.mecanum_move_xyz(0,0,20)
            time.sleep(0.1)
        else:
            got.mecanum_stop()
            break
            
    print("Apriltag centralized.")  

def AP_approaching():
    # Step 3 AP approach
    while True:
        AP_info = got.get_apriltag_total_info()
        if AP_info:
            distance =AP_info[0][6]
            x_coordinate = AP_info[0][1]
            if x_coordinate<300:
                got.mecanum_move_xyz(-5,5,0)
            elif x_coordinate>340:
                got.mecanum_move_xyz(5,5,0)
            elif distance>0.16:
                got.mecanum_move_xyz(0,10,0)
            else:
                got.mecanum_stop()
                break
    print("It's too close, stop!")
```
